# Mean dynamics

$$\dot{x} = p - q$$

- $p$: 
    - Probability of selecting a player without optimal strategy: (1-x) 
    - Probability of selecting an opponent with optimal strategy: x
    - Probability of getting payoff 0 at every test: x(1-x)
    - When the case above occurs, we must multiply by $\frac{1}{n}$
$$p = (1-x)\left[x + \frac{x(1-x)}{n}\right]$$
- $q$: 
    - Probability of selecting an player with optimal strategy: x 
    - Probability of getting payoff 0 when the agent tests the optimal strategy and payoff > 0 when testing any other strategy: $(1-x)^2$
    - Probability of getting payoff 0 at every test: x(1-x)
    - When the case above occurs, we must multiply by $\frac{n-1}{n}$
$$q = x\left[(1-x)^2 + \frac{x(1-x)(n-1)}{n}\right]$$
$$\dot{x} = (1-x)\left[x + \frac{x(1-x)}{2}\right] - x\left[(1-x)^2 + \frac{x(1-x)}{2}\right]$$

$$\dot{x} = (1-x)x\left[1 + \frac{(1-x)}{2}\right] - x(1-x)\left[(1-x) + \frac{x}{2}\right]$$

$$\dot{x} = (1-x)x\left[1 + \frac{(1-x)}{2} - (1-x) - \frac{x}{2}\right]$$

$$\dot{x} = (1-x)x\left[\frac{1}{2} -\frac{x}{2} + x - \frac{x}{2}\right] = \frac{(1-x)x}{2}$$

- Noise:
    - With probability (1 - noise) everything remain the same.
    - With probability (noise mutiply by 1/n), x may grows if the reviewing agent has non optimal strategy and vice versa:
        - $p_{noise} = (1-x)\cdot\frac{\epsilon}{2}$
        - $q_{noise} = x\cdot\frac{\epsilon}{2}$
$$\dot{x} = (1-\epsilon)\cdot\frac{(1-x)x}{2} + \epsilon\cdot(\frac{1}{2} - x)$$

```python
def mean_dynamics(x, t):
    return (1 - x) * x / 2
```

```python
def mean_dynamics_noise_1(x, t):
    noise = 0.1
    return (1 - noise) * ((1 - x) * x / 2) + noise * (0.5 - x)

def mean_dynamics_noise_01(x, t):
    noise = 0.01
    return (1 - noise) * ((1 - x) * x / 2) + noise * (0.5 - x)

def mean_dynamics_noise_001(x, t):
    noise = 0.001
    return (1 - noise) * ((1 - x) * x / 2) + noise * (0.5 - x)

def mean_dynamics_noise_05(x, t):
    noise = 0.05
    return (1 - noise) * ((1 - x) * x / 2) + noise * (0.5 - x)
```

```python
t = np.linspace(0, 100, 101)
initial_proportion_optimal_strategists = 0.1
```

```python
mean_dynamics_integration = odeint(mean_dynamics, initial_proportion_optimal_strategists, t)
```

```python
mean_dynamics_integration_1 = odeint(mean_dynamics_noise_1, initial_proportion_optimal_strategists, t)
mean_dynamics_integration_01 = odeint(mean_dynamics_noise_01, initial_proportion_optimal_strategists, t)
mean_dynamics_integration_01 = odeint(mean_dynamics_noise_01, initial_proportion_optimal_strategists, t)
mean_dynamics_integration_001 = odeint(mean_dynamics_noise_001, initial_proportion_optimal_strategists, t)
mean_dynamics_integration_05 = odeint(mean_dynamics_noise_05, initial_proportion_optimal_strategists, t)
```

```python
plt.plot(mean_dynamics_integration_1, label="MD")
plt.legend()
```
